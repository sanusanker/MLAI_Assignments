```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataOfDiabetes = pd.read_csv('diabetes.csv')
```

```python
#Display the columns in the dataset.
dataOfDiabetes.columns
```

```python
#Calculate the mean of each column of the dataset.
dataOfDiabetes_mean = dataOfDiabetes.mean()
```

```python
print(dataOfDiabetes_mean)
```

```python
#Check for the null values present in the dataset.
dataOfDiabetesOfNull = dataOfDiabetes.isna()
print(dataOfDiabetesOfNull)
count = dataOfDiabetesOfNull.sum()
print(count)
```

```python
# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations
plt.hist(dataOfDiabetes['Age'],color='aqua')
plt.title('Based on age')
plt.xlabel('age')
plt.ylabel('count')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataOfDiabetes['Age'],dataOfDiabetes['Glucose'])
plt.title("Scatterplot of Age vs Glucose")
plt.xlabel("Age")
plt.ylabel("Glucose")
```
